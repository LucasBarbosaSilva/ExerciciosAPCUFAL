1. Escreva uma função que conta a quantidade de
vogais em um texto e armazena tal quantidade em
um dicionário, onde a chave é a vogal considerada.

```python
import unicodedata

texto = input()

#normalizando o texto
textoNormalizado = unicodedata.normalize('NFD', texto)
texto = textoNormalizado.encode('ascii', 'ignore').decode('utf8').casefold()

vogais = ['a', 'e', 'i', 'o', 'u']
dicionarioVogais = {}
totalVogais = 0

for x in vogais:
  quantidade = texto.count(x)
  dicionarioVogais[x]= quantidade
  totalVogais += quantidade

print("O total de vogais encontradas foi de: ", totalVogais)
for key, value in dicionarioVogais.items():
  if(value == 1):
    print("A vogal", key, "apareceu", value, "vez" )
  else:
    print("A vogal", key, "apareceu", value, "vezes" )
```

2. Escreva um programa que lê duas notas de
vários alunos e armazena tais notas em um
dicionário, onde a chave é o nome do aluno. A
entrada de dados deve terminar quando for lida uma
string vazia como nome. Escreva uma função que
retorna a média do aluno, dado seu nome.

```python
def media(nome):
  return sum(notas[nome])/2

notas = dict()
nome = input("Digite o nome do aluno: ")
while(nome != ""):
  nota1 = float(input("Digite a primeira nota: "))
  nota2 = float(input("Digite a segunda nota: "))
  notas[nome] = [nota1, nota2]
  nome = input("Digite o nome do aluno: ")

aluno = input("De qual aluno você deseja ver a média? ")
print("A média do aluno", aluno, "é:", media(aluno))
```
